```python
import pandas as pd

df = pd.read_csv("QM 2 projects data - data for part 1.csv")
```

```python
# Summary statistics

columns = ["lower secondary school completion rate", "maternal mortality"]
df[columns].agg(["count", "mean", "std", "min", "max"]).round(2)
```

```python
# Boxplot for maternal mortality

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df, y="maternal mortality")
plt.ylabel("Maternal mortality (per 100,000 live births)")
plt.title("Boxplot of maternal mortality")
plt.show()
```

```python
# Boxplot for female lower secondary school completion rate

sns.boxplot(data=df, y="lower secondary school completion rate")
plt.ylabel("Female lower secondary school completion rate (%)")
plt.title("Boxplot of female lower secondary school completion rate")
plt.show()
```

```python
# Line graph of female lower secondary school completion rate and maternal mortality over time

x = df["year"]
y1 = df["lower secondary school completion rate"]
y2 = df["maternal mortality"]

fig, ax1 = plt.subplots()

line1, = ax1.plot(x, y1, color="red", label="Female lower secondary school completion rate (%)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Female lower secondary school completion rate (%)")
ax1.tick_params(axis="y")

ax2 = ax1.twinx()
line2, = ax2.plot(x, y2, color="purple", label="Maternal mortality")
ax2.set_ylabel("Maternal mortality (per 100,000 live births)")
ax2.tick_params(axis="y")

plt.title("Female lower secondary school completion rate and maternal mortality over time")


ax1.legend(handles=[line1, line2], loc="upper left")

ax1.set_xticks(x[::5])
ax1.set_xticklabels(x[::5], rotation=45)

plt.show()
```
